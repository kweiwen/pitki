```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define ranges
listn_2_center_spk_vals = np.linspace(0.4, 3.0, 270)
spk_2_spk_vals = np.linspace(0.1, 0.8, 80)
ear_2_ear = 0.18
wind_speed = 343
sample_rate = 48000
```

```python
# Create meshgrid
X, Y = np.meshgrid(listn_2_center_spk_vals, spk_2_spk_vals)
s = Y / 2
a = ear_2_ear / 2

# Calculate distances
d_ipsi = np.sqrt(X**2 + (a - s) ** 2)
d_cont = np.sqrt(X**2 + (a + s) ** 2)
d_diff = np.abs(d_cont - d_ipsi)
taosamples = d_diff * sample_rate / wind_speed
taosamples_int = np.round(taosamples).astype(int)
```

```python
# Plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, taosamples_int, levels=19, cmap='viridis')
plt.colorbar(contour, label='|d_cont - d_ipsi| (m)')
plt.xlabel('listn_2_center_spk (m)')
plt.ylabel('spk_2_spk (m)')
plt.title('Tao Samples')
plt.grid(True)
plt.show()
```
