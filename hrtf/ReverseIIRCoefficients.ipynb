```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz, sosfilt, sosfreqz, bilinear
import math
```

```python
b = np.array([-0.25, 0.75])
a = np.array([1.0, -0.5])
w, h1 = freqz(b, a)
_, h2 = freqz(np.flip(b), np.flip(a))
_, h3 = freqz(np.convolve(b, np.flip(b)), np.convolve(a, np.flip(a)))
```

```python
amp1 = 20 * np.log10(abs(h1))
ang1 = np.angle(h1)
amp2 = 20 * np.log10(abs(h2))
ang2 = np.angle(h2)
amp3 = 20 * np.log10(abs(h3))
ang3 = np.angle(h3)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

plt.title('frequency response of filter')
plt.xlabel(r'normalized frequency (x$\pi$rad/sample)')

ax1.plot(w/max(w), amp1, label='1')
ax1.plot(w/max(w), amp2, label='2')
ax1.plot(w/max(w), amp3, label='3')
ax1.set_ylabel('amplitude (dB)')
ax1.set_ylim(-24, 24)
ax1.grid()

ax2.plot(w/max(w), ang1, label='1')
ax2.plot(w/max(w), ang2, label='2')
ax2.plot(w/max(w), ang3, label='3')
ax2.set_ylabel('phase (radians)')
ax2.set_ylim(-2, 2)

ax1.legend()
plt.xscale("log")
plt.show()
```
